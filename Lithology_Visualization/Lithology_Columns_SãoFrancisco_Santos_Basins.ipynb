```python
# Import required dependencies 
import numpy as np
import pandas as pd
import matplotlib.colors as colors
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt                                               
from matplotlib_inline.backend_inline import set_matplotlib_formats 
set_matplotlib_formats('svg')
```

```python
# Create the lithological column plot 
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(25,26))
fig.subplots_adjust(top=0.8, wspace=0.15)

# Set up subplots grid
ax1 = plt.subplot2grid((1,4),(0,0),rowspan=1,colspan=1)  
ax2 = plt.subplot2grid((1,4),(0,1),rowspan=1,colspan=1)  
ax3 = plt.subplot2grid((1,4),(0,2),rowspan=1,colspan=1) 
ax4 = plt.subplot2grid((1,4),(0,3),rowspan=1,colspan=1) 

####################################################################################################

# Define the lithology data
depth_intervals1   = [(3241,3300), (3300,3350), (3350,3360),(3360,3365),(3365,3370),(3370,3372),(3372,3375), (3375,3380), (3380,3385), (3385,3400),(3400,3403), (3403,3415)]
lithology_colors1  = ['dodgerblue', 'yellow', 'yellow', 'dodgerblue','yellow','dodgerblue','yellow','lime','crimson','lime','crimson','lime']
lithology_hatches1 = ['-/','-.','-.','-/','-.','-/','-.','--','-...','--','-...','--']

# Customize the plot
ax1.set_xlim(0,1)
ax1.set_ylim(3415,3241)  
ax1.set_xlabel('Lithology', color='black', fontsize=13) 
ax1.set_ylabel('Depth1 (m)', color='black', fontsize=13) 
ax1.set_title('Well 1-BRSA-871-MG', color='black', fontsize=16, fontweight='bold')

for depth, color, hatch in zip(depth_intervals1, lithology_colors1, lithology_hatches1):
    ax1.fill_betweenx([depth[0], depth[1]], 0, 1, color=color, hatch=hatch, edgecolor='black', lw=0.5) 
    
####################################################################################################    
    
# Define the lithology data
depth_intervals2   = [(3250,3290), (3290,3300), (3300,3310), (3310,3320), (3320,3330), (3330,3600)]
lithology_colors2  = ['lime', 'crimson', 'lime', 'yellow','red', 'yellow']
lithology_hatches2 = ['--','-...','--','-.','+++','-.']

# Customize the plot
ax2.set_xlim(0,1)
ax2.set_ylim(3600,3250)             
ax2.set_xlabel('Lithology', color='black', fontsize=13) 
ax2.set_ylabel('Depth2 (m)', color='black',fontsize=13) 
ax2.set_title('Well 1-BRSA-948-MG', color='black', fontsize=16, fontweight='bold')

for depth, color,hatch in zip(depth_intervals2, lithology_colors2, lithology_hatches2):
    ax2.fill_betweenx([depth[0], depth[1]], 0, 1, color=color, hatch=hatch, edgecolor='black', lw=0.5)      
                    
####################################################################################################                     
    
# Define the lithology data
depth_intervals3   = [(4975,5340), (5340,5360), (5360,5790), (5790,5800)]
lithology_colors3  = ['cornflowerblue', 'magenta', 'paleturquoise', 'lime']
lithology_hatches3 = ['x','/','+','--']

# Customize the plot
ax3.set_xlim(0,1)
ax3.set_ylim(5800,4975)         
ax3.set_xlabel('Lithology', color='black', fontsize=13) 
ax3.set_ylabel('Depth3 (m)', color='black', fontsize=13) 
ax3.set_title('Well 1-BRSA-1116-RJS', color='black', fontsize=16, fontweight='bold')

for depth, color, hatch in zip(depth_intervals3, lithology_colors3, lithology_hatches3):
    ax3.fill_betweenx([depth[0], depth[1]], 0, 1, color=color, hatch=hatch, edgecolor='black', lw=0.5)
    
####################################################################################################
    
# Define the lithology data
depth_intervals4   = [(5342,5400), (5400,5415), (5415,5635), (5635,5650), (5650,5684)]
lithology_colors4  = ['cornflowerblue', 'magenta', 'paleturquoise', 'lime', 'paleturquoise']
lithology_hatches4 = ['x','/','+','--','+']

# Customize the plot
ax4.set_xlim(0,1)
ax4.set_ylim(5684,5342)             
ax4.set_xlabel('Lithology', color='black', fontsize=13) 
ax4.set_ylabel('Depth4 (m)', color='black', fontsize=13) 
ax4.set_title('Well 3-BRSA-1215-RJS', color='black', fontsize=16, fontweight='bold')

for depth, color, hatch in zip(depth_intervals4, lithology_colors4, lithology_hatches4):
    ax4.fill_betweenx([depth[0], depth[1]], 0, 1, color=color, hatch=hatch, edgecolor='black', lw=0.5)
    
####################################################################################################

# Reduce the space between each subplot
fig.subplots_adjust(wspace = 0.5)
```

```python
# Create Lithology Legend for So Francisco Basin

fig, ax = plt.subplots(figsize=(5,5))
fig.suptitle("Lithology So Francisco Basin", fontsize=20, fontweight='bold')
fig.subplots_adjust(top=0.8)

# Define the lithology data
lithology_labels  = ['Dolomite','Igneous','Limestone','Sandstone','Shale','Silt']
lithology_colors  = ['dodgerblue','red','paleturquoise','yellow','lime','crimson']
lithology_hatches = ['-/','+++','+','-.','--','-...']

# Create the legend patches
legend_patches = []
for color, hatch, label in zip(lithology_colors, lithology_hatches, lithology_labels):
    patch = mpatches.Patch(facecolor=color, hatch=hatch, edgecolor='black', label=label)
    legend_patches.append(patch)

# Create the legend
legend = plt.legend(handles=legend_patches, fontsize=20, framealpha=1, loc='center', labelspacing=1.5)
legend.get_frame().set_linewidth(1.8)
legend.get_frame().set_edgecolor('black')
plt.axis('off')
```
