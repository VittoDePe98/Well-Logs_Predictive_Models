<span style='color:black'> <span style="font-size:40px;"> **WELL 1-BRSA-871-MG**

<span style='color:deepskyblue'> <span style="font-size:19px;"> **Loading the Well Log Data**

```python
# Create variable las containing all the well log data 
# "/Users/vittoriadepellegrini/Desktop/Tesi PoliTO/1-brsa-871-mg_8_tdd_cnl_ait_converted_with_python_1.las 2" = directory + file name
# The uploaded file was originally a DLIS file which has been converted to a LAS file
las = lasio.read("/Users/vittoriadepellegrini/Desktop/Tesi PoliTO/1-brsa-871-mg_8_tdd_cnl_ait_converted_with_python_1.las 2")
```

```python
# Load the well log data to Pandas DataFrame 
df = las.df()
df.head(12781)  # df.head() returns the rows of the DataFrame; in this case, 12781 is the total number of rows
```

```python
# Create a "column Index" within the DataFrame
df.reset_index(inplace=True)  # inplace=True is set to modify the original DataFrame
df.head(12781)
```

```python
# Rename the column TDEP to DEPTH
df.rename(columns={'TDEP':'DEPTH'}, inplace=True)
df.head(12781)
```

```python
# Visualize the MIN and MAX values for each well log measurement 
df.describe()
```

```python
# Visualize the type of well log data 
df.info()
```

```python
# Convert DEPTH values to metres, by multiplying by 0.00254. 
# Add a new column to the DataFrame, called DEPTH1.

df['DEPTH1'] = " "                  # add column DEPTH1
new_column = df.pop('DEPTH1')       # shift column DEPTH1 to second position
df.insert(1,'DEPTH1', new_column)   # insert column using insert(position, column_name, new_column) function
df.DEPTH1 = 0.00254*df.DEPTH        # convert depth from inches into metres
df.head(12781)
```

```python
# Convert the (DTCO3) Sonic Compressional Slowness (us/ft) into a Sonic Compressional Velocity (m/s). 
# Add a new column to the DataFrame, called Vp (m/s). 
# 1 ft = 0.3048 m
# 1 microsec = 1e-6 sec

df['Vp'] = ""
new_column1 = df.pop('Vp')
df.insert(2, 'Vp', new_column1)
df.Vp = ((df.DTCO3)**(-1))*(0.3048/1e-6)
df.head(12781)
```

```python
pd.set_option("display.max.columns", None)  # To make sure that Pandas does not hide any column 
df.describe()
```

```python
# Visualize only the columns of interest 
df[['DEPTH1', 'GR', 'NPHI', 'RHGX_HILT', 'PHIE_HILT', 'RT', 'RWA_HILT']] 
```

**Remove all the NON-SENSE values (DATA PRE-PROCESSING):**

```python
# HILT Effective Porosity: remove all the 0 values 
df.drop(df.loc[df['PHIE_HILT']==0].index, inplace=True)
```

```python
# HILT Grain Density: remove all the negative values 
df.drop(df.loc[(df['RHGX_HILT'] < 0)].index, inplace=True)
```

```python
# Removing the NaN values from the dataset (NaN = not a number)
df = df.dropna()
```

```python
# Final DataFrame 
df.describe()
```

```python
df.info()
```

```python
# Convert a DataFrame into a CSV file (CSV file is required for ML Algorithms implementation)
df.to_csv('well_log_data.csv', index=None)  # well_log_data.csv is the name of the csv file
```
